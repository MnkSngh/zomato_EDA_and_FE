## Next Data Set

```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

#### Combining Two Data Sets

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left') ## df and df_country has common column 'country code' 
                                                                        ## basis which these two dataframes can be combined
```

```python
final_df.head(2)
```

## Rating Analysis

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
## to conert into DF and renaming unknown column
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating text',data = ratings, palette = ['blue','red','orange','yellow','pink','green'])
```
